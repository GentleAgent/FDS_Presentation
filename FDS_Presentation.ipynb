```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = 'movies_data.csv'
movies_df = pd.read_csv(file_path, encoding='ISO-8859-1')


movies_df['Budget'] = pd.to_numeric(movies_df['Budget'], errors='coerce')
movies_df['Box Office'] = pd.to_numeric(movies_df['Box Office'], errors='coerce')
movies_df['IMDb score'] = pd.to_numeric(movies_df['IMDb score'], errors='coerce')
```

```

plt.figure(figsize=(8, 6))
correlation_matrix = movies_df[['Budget', 'Box Office', 'IMDb score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Budget, Box Office, and IMDb Score')
plt.show()
```

```

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Budget', y='Box Office', data=movies_df)
plt.title('Box Office Revenue vs Budget')
plt.xlabel('Budget ($)')
plt.ylabel('Box Office Revenue ($)')
plt.grid(True)
plt.show()
```

```

genre_year_distribution = movies_df.groupby(['Release year', 'Genre']).size().unstack().fillna(0)


plt.figure(figsize=(12, 8))
genre_year_distribution.plot(kind='line', stacked=False)
plt.title('Trends in Movie Genres Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(8, 6))
sns.histplot(movies_df['IMDb score'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of IMDb Scores')
plt.xlabel('IMDb Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```
# Nominations vs Awards
plt.figure(figsize=(8, 6))
movies_df[['Oscar and Golden Globes nominations', 'Oscar and Golden Globes awards']].sum().plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Oscar and Golden Globes: Nominations vs Awards')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```
# Box plot for Box Office revenue by Genre
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genre', y='Box Office', data=movies_df)
plt.title('Box Office Earnings by Genre')
plt.xticks(rotation=45)
plt.ylabel('Box Office Revenue ($)')
plt.show()
```

```
# Average IMDb score by Genre
avg_imdb_by_genre = movies_df.groupby('Genre')['IMDb score'].mean().sort_values()

# Plotting average IMDb score by genre
plt.figure(figsize=(10, 6))
avg_imdb_by_genre.plot(kind='barh', color='lightgreen')
plt.title('Average IMDb Score by Genre')
plt.xlabel('IMDb Score')
plt.ylabel('Genre')
plt.grid(True)
plt.show()
```
